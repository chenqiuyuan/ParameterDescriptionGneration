### 2.4（完整数据：valid/test/train）原始数据 -> 函数参数+注释参数

````python
import pandas as pd
"""批量抽取valid/test/train"""



def batch_extract_signature(raw_file_dict, signature_file_dict):
    mode_list = ["valid", "test", "train"]
    for mode in mode_list:
        explore_data.extract_signature_from_file(raw_file_dict[mode], signature_file_dict[mode])

def batch_extract_param_doc(raw_file_dict, param_doc_file_dict):
    mode_list = ["valid", "test", "train"]
    for mode in mode_list:
        explore_data.extract_docstring_params_from_file(raw_file_dict[mode], param_doc_file_dict[mode])

def batch_merge_output(signature_file_dict, param_doc_file_dict, output_file_dict):
    mode_list = ["valid", "test", "train"]
    for mode in mode_list:
        sig = pd.read_json(signature_file_dict[mode], lines=True) 
        param = pd.read_json(param_doc_file_dict[mode], lines=True)

        merge = pd.concat([sig, param], axis=1)

        merge.to_json(output_file_dict[mode], orient="records", lines=True)

# batch_extract_signature(raw_file_dict, signature_file_dict)
# batch_extract_param_doc(raw_file_dict, param_doc_file_dict)
batch_merge_output(signature_file_dict, param_doc_file_dict, output_file_dict)
````

### 2.5 （完整数据：valid/test/train） 原始数据->(函数粒度：每个函数+参数名列表+参数注释列表+参数数据流)

数据格式：
raw_data -> <parameter, code, parameter flow，@parameter>

````python
import pandas as pd
from IPython.display import display

valid_raw = pd.read_json(raw_file_dict["valid"], lines=True)
test_raw = pd.read_json(raw_file_dict["test"], lines=True)
train_raw = pd.read_json(raw_file_dict["train"], lines=True)


valid_param_info = pd.read_json(param_info_dict["valid"], lines=True)
test_param_info = pd.read_json(param_info_dict["test"], lines=True)
train_param_info = pd.read_json(param_info_dict["train"], lines=True)

valid_param_info = valid_param_info[["param_list", "param_doc", "param_flow"]].dropna(how="any")
test_param_info = test_param_info[["param_list", "param_doc", "param_flow"]].dropna(how="any")
train_param_info = train_param_info[["param_list", "param_doc", "param_flow"]].dropna(how="any")

display(valid_param_info)
display(test_param_info)
display(train_param_info)

````

````python
valid_func = valid_raw[["func_name", "code"]].iloc[valid_param_info.index]
test_func = test_raw[["func_name", "code"]].iloc[test_param_info.index]
train_func = train_raw[["func_name", "code"]].iloc[train_param_info.index]


valid_output = pd.concat([valid_func, valid_param_info], axis=1)
test_output = pd.concat([test_func, test_param_info], axis=1)
train_output = pd.concat([train_func, train_param_info], axis=1)

# https://stackoverflow.com/questions/39011511/pandas-expand-rows-from-list-data-available-in-column
# Pandas 的explode可以直接用

valid_output = valid_output.explode(["param_list", "param_doc", "param_flow"]).dropna(how="any")
test_output = test_output.explode(["param_list", "param_doc", "param_flow"]).dropna(how="any")
train_output = train_output.explode(["param_list", "param_doc", "param_flow"]).dropna(how="any")

valid_output = valid_output.rename_axis("index_in_raw").reset_index()
test_output = test_output.rename_axis("index_in_raw").reset_index()
train_output = train_output.rename_axis("index_in_raw").reset_index()

display(valid_output)
display(test_output)
display(train_output)

valid_output.to_json(output_file_dict["valid"], orient="records", lines=True)
test_output.to_json(output_file_dict["test"], orient="records", lines=True)
train_output.to_json(output_file_dict["train"], orient="records", lines=True)
````

### 4.2 统计相同的parameter flow有多少

````python
import pandas as pd

CODESEARCHNET = "/data/ncc_data/code_search_net_feng"

output_file_dict = {
    "valid": "",
    "test": "",
    "train": ""
}
mode_list = ["valid", "test", "train"]
lang_list = ["java"]

mode_list = ["valid", "test", "train"]
lang_list = ["java"]
for lang in lang_list:
    for mode in mode_list:
        output_file_dict[mode] = os.path.join(CODESEARCHNET, "param_generation", lang, "{}.jsonl".format(mode))

check_valid = pd.read_json(output_file_dict["valid"], lines=True)
check_test = pd.read_json(output_file_dict["test"], lines=True)
check_train = pd.read_json(output_file_dict["train"], lines=True)
````

````python
dup_flow_valid = check_valid[check_valid["param_flow"].duplicated()]
dup_flow_test = check_test[check_test["param_flow"].duplicated()]
dup_flow_train = check_train[check_train["param_flow"].duplicated()]
````

````python
check_valid["code_len"] = check_valid["code"].apply(lambda x: len(x.split("\n")))
check_test["code_len"] = check_test["code"].apply(lambda x: len(x.split("\n")))
check_train["code_len"] = check_train["code"].apply(lambda x: len(x.split("\n")))

dup_flow_valid = check_valid[(check_valid["code_len"] > 10) & (check_valid["param_flow"].duplicated())]
dup_flow_test = check_test[(check_test["code_len"] > 10) & (check_test["param_flow"].duplicated())]
dup_flow_train = check_train[(check_train["code_len"] > 10) & (check_train["param_flow"].duplicated())]

display(len(dup_flow_valid))
display(len(dup_flow_test))
display(len(dup_flow_train))
````

````python
display(len(dup_flow_valid) / len(check_valid))
display(len(dup_flow_test) / len(check_test))
display(len(dup_flow_train) / len(check_train))
````
